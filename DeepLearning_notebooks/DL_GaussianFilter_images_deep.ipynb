## Example of Gaussian filter for any image

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

import numpy as np
```

```python
from PIL import Image
import requests
import numpy as np
from StringIO import StringIO

response = requests.get('https://c2.staticflickr.com/2/1678/26387815032_9a0d618045_b.jpg')
img = np.array(Image.open(StringIO(response.content)))

img[0:2]
```

```python
plt.imshow(img)
```

```python
#import cv2
from scipy.ndimage.filters import gaussian_filter
import scipy
```

```python

x_gf = gaussian_filter(img, 5)
x_resized = scipy.misc.imresize(x_gf, (int(0.25 * img.shape[0]), int(0.25 * img.shape[1])))
```

```python
plt.imshow(x_resized)
```
