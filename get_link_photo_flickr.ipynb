# Get links from Flickr 

```python

from bs4 import BeautifulSoup
import urllib2
url = 'https://www.flickr.com/photos/marcelamcgreal/26387815032/'

def get_link(url_param):
    pageFile = urllib2.urlopen(str(url_param))
    pageHtml = pageFile.read()
    pageFile.close()
    soup = BeautifulSoup("".join(pageHtml),'lxml')
    a = soup.find_all('meta')
    for num in range(0,len(a)):
        #print num,a[num]
        #print str(a[num])
        if 'staticflickr' in str(a[num]):
            link = a[num]["content"]
    return link

result = get_link(url)

print "link", result
```
